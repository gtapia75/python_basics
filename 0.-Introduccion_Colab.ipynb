# Ejemplos de Code Cell

en los notbooks los code cell son basicamente las celdas en donde podemos escribir y ejecutar comandos de Python. Aqui daremos algunos ejemplos al respecto

```python
# Comentarios inician con el #
```

```python
# se pueden ejecutar comandos e imprimirlos directamente
5 + 5
```

```python
# Se ejecutan todos pero solo el ultimo es presentado
5 + 5
10 + 10
```

```python
# Para imprimir todos los resultados se puede usar la funcion print
print(5+5)
print(10+10)
```

```python
# Se pueden imprimir strings
print("Hola Mundo")
```

```python
# Se pueden concatenar strings
print("Hola" + "Mundo")
print("Hola" + " " + "Mundo")
# se puede hacer append de strings, agregando por default un espacio entre ellos
print("Hola", "Mundo")
print("Hola", "Mundo", sep="")
```

```python
# No se pueden concatenar distintos tipos de datos
print("Hola" + 5)
```

```python
# pero se puede hacer de esta manera
print("Hola ", str(5))
```

```python
# Se pueden imprimir variables
string1 = "Hola"
string2 = "Mundo"
print(string1 + " " + string2)
```

```python
#Se pueden formatear lo que se quiere imprimir
int1=5
print(f"Hola {int1}")
```
