# Tipos de Datos

### Tipo de datos Numericos

*   `int` Valores enteros que no contienen punto decimal
*   `float` Valores que contienen punto decimal y aproix 15 decimales de precision
*   `complex` Numeros que representan 3+4j donde j es la parte imaginaria

### Tipo de datos de secuencia

*   `str` Son tipo string que pueden usar comilla simple o doble
*   `list` Secuencias ordenadas, mutables [1,2,3] ['1','2','3'] pueden contener distintos tipos [1,'dos', 3.0]. Uso de []
*   `tuple` Secuencias ordenadas inmutables (1,2,3) similares a las list pero que no pueden ser modificadas despues de ser creadas. Uso de ()
*   `range` Representan secuencias de numeros generalmente utilizadas para loops por ejemplo range(5) contendra los numeros 0,1,2,3,4

### Tipo de datos Mapeo

*   `dict` Pares Key-Value donde las keys deben ser inmutables, {'nombre':'Juan', 'edad':40}

### Tipo de datos Set

*   `set` Colleccion sin orden de elementos unicos, {1,2,3}

### Tipo de datos Boolean

*   `bool` Acepta valores True o False, utilizado para operaciones logicas, case sensitive

### Tipo de datos None

*   `None` Representa la ausencia de un valor o un valor null


```python
# Ejemplos numericos
edad = 25
altura = 1.75
num_complejo = 3 + 4j
print(edad, type(edad))
print(altura, type(altura))
print(num_complejo, type(num_complejo))
```

```python
# Ejemplo de secuencias
nombre = "Alicia"
numeros = [1,2,3]
coordenadas = (10,20)
print(nombre, type(nombre))
print(numeros, type(numeros))
print(coordenadas, type(coordenadas))
```

```python
# Ejemplo de Mapeo o Dict
persona = {
    "nombre": "Alicia",
    "edad": 25,
    "altura": 1.75
}
print(persona, type(persona))
```

```python
# Ejemplo de Sets
numeros_unicos = {1,2,3,1,2}
print(numeros_unicos, type(numeros_unicos))
```

```python
# Ejemplo boolean
is_active = True
print(is_active, type(is_active))
```

```python
# Ejemplo None
value = None
print(value, type(value))
```

## Conversiones de Datos
Python soporta algunos tipos de conversiones de datos, tambien llamado type casting

*   Conversiones a String
*   Conversiones a Int
*   Conversiones a float
*   Conversiones a boolean



```python
#  Conversiones a String
num_str = str(123)        # Integer a string: "123"
float_str = str(3.14)     # Float a string: "3.14"
bool_str = str(True)      # Boolean a string: "True"

# Conversiones a Int
str_num = int("123")      # String a integer: 123
float_num = int(3.14)     # Float a integer: 3 (trunca la parte decimal)
bool_num = int(True)      # Boolean a integer: 1 (True=1, False=0)

# Conversiones a Float
str_float = float("3.14") # String a float: 3.14
int_float = float(123)    # Integer a float: 123.0
bool_float = float(True)  # Boolean a float: 1.0

# Conversiones a Boolean
str_bool = bool("True")   # String a boolean: True
num_bool = bool(1)        # Number a boolean: True (0 es False, todos los demas numeros son True)
float_bool = bool(0.0)    # Float a boolean: False
```

### Conversiones en colecciones
Python tambien soporta hacer casting a distintos tipos

*   Conversiones a list
*   Conversiones a tuple
*   Conversiones a set
*   Conversion a dict


```python
# Conversiones a List
tuple_list = list((1, 2, 3))     # Tuple a list: [1, 2, 3]
set_list = list({1, 2, 3})       # Set a list: [1, 2, 3]
str_list = list("Python")        # String a list: ['P', 'y', 't', 'h', 'o', 'n']

# Conversiones a tuple
list_tuple = tuple([1, 2, 3])    # List a tuple: (1, 2, 3)
set_tuple = tuple({1, 2, 3})     # Set a tuple: (1, 2, 3)
str_tuple = tuple("Python")      # String a tuple: ('P', 'y', 't', 'h', 'o', 'n')

# Conversiones a Set
list_set = set([1, 2, 2, 3])     # List a set: {1, 2, 3} (Elimina los  duplicados)
tuple_set = set((1, 2, 2, 3))    # Tuple a set: {1, 2, 3}
str_set = set("Python")          # String a set: {'P', 'y', 't', 'h', 'o', 'n'}

# Conversiones a Dictionary
list_dict = dict([[1, 'uno'], [2, 'dos']])  # List de pares a dict: {1: 'uno', 2: 'dos'}
tuple_dict = dict(((1, 'uno'), (2, 'dos'))) # Tuple de pares a dict: {1: 'uno', 2: 'dos'}
```

## Operadores Arimeticos Basicos

*   Adicion `+` `5 + 3 = 8`
*   Sustraccion `-` `5 - 3 = 2 `
*   Multiplicacion `*` `5 * 3 = 15`
*   Division `/` `5 / 3 = 1.66`
*   Division Entera `//` `5 // 3 = 1` Divide y redondea hacia abajo al entero cercano
*   Modulo `%` `5 % 3 = 2` Regr[link text](https://)esa el residuo de la division
*   Exponenciacion `**` `5 ** 3 = 125`



```python
x = 10
y = 3

# Operaciones Basicas
print(x + y)    # 13
print(x - y)    # 7
print(x * y)    # 30
print(x / y)    # 3.3333...
print(x // y)   # 3
print(x % y)    # 1
print(x ** y)   # 1000
```

## Operadores de Asignacion

Python tambien soporta operadores de asignacion como

*   `+=` Suma y asigna
*   `-=` Resta y asigna
*   `*=` Multiplica y asigna
*   `/=` Divide y asigna
*   `//=` Division Entera y asigna
*   `%=` Modulo y asigna
*   `**=` Potencia y asigna



```python
# Suma (+=)
x = 5
x += 3      # Equivalente a: x = x + 3
print(f'Resultado (+=): {x}')    # Resultado: 8

# Resta (-=)
x = 10
x -= 4      # Equivalente a: x = x - 4
print(f'Resultado (-=): {x}')    # Resultado: 6

# Multiplicacion (*=)
x = 3
x *= 4      # Equivalente a: x = x * 4
print(f'Resultado (*=): {x}')    # Resultado: 12

# Division (/=)
x = 15
x /= 3      # Equivalente a: x = x / 3
print(f'Resultado (/=): {x}')    # Resultado: 5.0

# Division Entera (//=)
x = 15
x //= 4     # Equivalente a: x = x // 4
print(f'Resultado (//=): {x}')    # Resultado: 3

# Modulo (%=)
x = 17
x %= 5      # Equivalente a: x = x % 5
print(f'Resultado (%=): {x}')    # Resultado: 2

# Exponenciacion (**=)
x = 2
x **= 3     # Equivalente a: x = x ** 3
print(f'Resultado (**=): {x}')    # Resultado: 8
```

```python
# Se pueden realizar operaciones con strings

# Concatenacion de String +=
saludo = "Hola"
saludo += " Mundo"   #  Equivalente a: greeting = greeting + " World"
print(saludo)        # Output: Hello World

# Multiplicacion de String *=
simbolo = "*"
simbolo *= 3            #  Equivalente a: symbol = symbol * 3
print(simbolo)          # Resultado: ***
```

```python
# Tambien se pueden ejecutar asignacion con listas

# Concatenacion de Listas (+=)
numeros = [1,2,3]
numeros += [4,5] # Equivalente a: numeros = numeros + [4,5]
print(f'Listas con (+=): {numeros}')

# Multiplicacion de listas
numeros *= 2     # Equivalente a: numeros = numeros * 2
print(f'Listas con (*=): {numeros}')
```

## Operadores de Comparacion

Los operadores de comparacion son muy similares a cualquier otro lenguaje de programacopm

*   `==` Igual a
*   `!=` Diferente
*   `>` Mayor que
*   `<` Menor que
*   `>=` Mayor igual que
*   `<=` Menor igual que



```python
# Igual a (==)
x = 5
y = 5
z = "5"
print(x == y)      # True (Comparando 2 numeros)
print(x == z)      # False (Comparando numero vs string)
print(x == int(z)) # True (Comparando usando conversaion)

# Diferente a (!=)
a = 10
b = 20
print(a != b)      # True
print(5 != 5)      # False

# Mayor que (>)
print(10 > 5)      # True
print(5 > 5)       # False
print(3 > 7)       # False

# Menor que (<)
print(5 < 10)      # True
print(5 < 5)       # False
print(7 < 3)       # False

# Mayor igual a (>=)
print(10 >= 5)     # True
print(5 >= 5)      # True
print(3 >= 7)      # False

# Menor igual a (<=)
print(5 <= 10)     # True
print(5 <= 5)      # True
print(7 <= 3)      # False
```

```python
# Algunos ejemplos adicionales

# Validacion de Password
stored_password = "secret123"
user_input = "secret123"
is_correct = stored_password == user_input
print(f"Password correcto? {is_correct}")  # True

# Calificaciones
calificacion = 85
calificacion_A = 90 <= calificacion <= 100
calificacion_B = 80 <= calificacion < 90
print(f"Resultado A?" , calificacion_A)    # False
print(f"Resultado B? {calificacion_B}")    # True
```

## Operadores Logicos

Python soporta los siguientes operadores Logicos


*   `and`
*   `or`
*   `not`



```python
# Operador AND (and)
# Regresa True si ambas condiciones son True
x = 5
y = 10

print(x > 0 and y < 20)     # True (Ambas condiciones son True)
print(x < 0 and y < 20)     # False (Primer condicion es False)
print(x > 0 and y > 20)     # False (Segunda condicion False)

# Operador OR (or)
# Regresa True if si al menos una de las condiciones True
edad = 25
tiene_id = False

print(edad >= 21 or tiene_id)  # True (Primera condicion es True)
print(edad < 21 or tiene_id)   # False (Ambas son False)

# Operador NOT (not)
# Invierte el valor de un boolean
esta_soleado = True
print(not esta_soleado)    # False
print(not (x > y))         # True (Porque x > y es False)
```

### Operadores de Identidad (Identity) o Membresia (Membership)

*   `is, is not` Operadores de Identidad
  *   Estos operadores comparan localidades de memoria de los objetos
*   `in, not in` Operadores de Membresia
  *   Estos operadores validan si un valor existe en una secuencia



```python
# Operadores de identidad

# is - Regresa True si ambas variables apuntan al mismo objeto
# is not - Regresa True si las variables apuntan a diferentes objetos

# Ejemplos con listas
lista1 = [1, 2, 3]
lista2 = [1, 2, 3]
lista3 = lista1

print(lista1 is lista3)      # True (mismo objeto)
print(lista1 is lista2)      # False (Diferentes objectos con los mismos valores)
print(lista1 is not lista2)  # True

# Ejemplo con tipo de dato None
x = None
print(x is None)           # True
print(x is not None)       # False

# Ejemplo con integers (nota: Python funciona con valores pequenos)
a = 5
b = 5
print(a is b)              # True

# Pero con numeros grandes:
x = 257
y = 257
print(x is y)              # False (different objects)
```

```python
# Operadores de Membresia

frutas = ['manzana', 'platano', 'naranja']
print('manzana' in frutas)        # True
print('mango' in frutas)        # False
print('mango' not in frutas)    # True

# String membership
mensaje = "Hola Mundo"
print('H' in mensaje)           # True
print('hello' in mensaje)       # False (case sensitive)
print('x' not in mensaje)       # True

# Dictionary membership (valida las keys)
persona = {'nombre': 'Alice', 'edad': 25}
print('nombre' in persona)         # True (checks keys)
print('Alice' in persona)        # False (not checking values)
print('direccion' not in persona)  # True
```

## Libreria MATH

Como hemos comentado existen librerias pre-configuradas que pueden ser utilizadas en Python, una de ellas es la libreria math

Funciones Matematicas Comunes
*   `sqrt` Raiz cuadrada
*   `ceil` Techo de un numero o interno superior inmediato
*   `floor` Piso de un numero o interno inferior inmediato
*   `factorial` Regresa el factorial de un numero
*   `gcd` Greatest Common Divisor o mayor numero divisor

Funciones Trigonometricas
*   `sin` Funcion Seno
*   `cos` Funcion Coseno
*   `tan` Funcion Tangente

Numeros constantes
*   pi Numero Pi
*   e Numero Euler



```python
import math

# Common mathematical functions
print(f'Raiz Cuadrada de 16 es: {math.sqrt(16)}')          # Square root: 4.0
print(f'El ceiling o techo del numero 3.2 es: {math.ceil(3.2)}')         # Ceiling: 4
print(f'El floor o piso del numero 3.2 es: {math.floor(3.2)}')        # Floor: 3
print(f'El factorial de 120 es: {math.factorial(5)}')      # Factorial: 120
print(f'El numero divisor mayor de 12 y 8 es: {math.gcd(12, 8)}')        # Greatest common divisor: 4

# Trigonometric functions (input in radians)
print(f'El seno de Pi/2 es: {math.sin(math.pi/2)}')    # Sine: 1.0
print(f'El Coseno de 0 es: {math.cos(0)}')            # Cosine: 1.0
print(f'La Tangente de Pi/4 es: {math.tan(math.pi/4)}')    # Tangent: ~1.0

# Constants
print(f'El numero Pi es: {math.pi}')                # π: 3.141592...
print(f'El numero Euler es: {math.e}')                 # e: 2.718281...
```
